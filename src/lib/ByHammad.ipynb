```python

from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
import re

import seaborn as sns
import pandas as pd
import numpy as np
import datetime
import pandas as pd
import timeit
import pickle, glob 
import time 

import gpxpy.geo
import random 
from operator import is_not
from functools import partial

from sklearn.externals import joblib
from sklearn.neural_network import MLPRegressor 
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder



import json 


# In[2]:


from pymongo import MongoClient
client = MongoClient('localhost:27017')
db_air_quality = client.air_quality_model_hkust


# In[3]:
```

```python
month = 60*60*24*60
number_months = 24
ts = time.time() - number_months * month
query = {"time": {"$gte": ts}}
print(query)

time_diff = 2*30*70 # 70 minutes 60 +10 buffer 
distance_diff = 7000 # 10 km
```

```python
data_q = db_air_quality.aqi_aggregation_hkust_5000_1800_subhour.find(query)
data_lis = []
for d in data_q:
    data_ ={}
    for k, v in d.items():
        if type(v) == type(dict()):
            for k1, v1 in v.items():
                try:
                    data_[k1] = v1['obs']
                except:
                    data_[k1] = v1
        else:
            try:
                data_[k] = v['obs']
            except:
                data_[k] = v
            
            
    data_lis.append(data_)
    
```

```python
df_air_quality = pd.DataFrame(data_lis)
df_air_quality = df_air_quality.drop('_id',1)
```

```python
spatial_features = ['latitude', 'longitude', 'station_code', 'station_name']

data_w = list(db_air_quality.weather_model_hkust.find(query))
df_weather = pd.DataFrame(data_w)
df_weather['longitude'] = df_weather['loc'].apply(lambda x :x[0])
df_weather['latitude'] = df_weather['loc'].apply(lambda x :x[1])
df_weather = df_weather.drop('loc',1)
df_weather = df_weather.drop('_id',1)
```

```python
def interpolate_AQI(row, df_tmp):
    
    aqui_features = ['AQHI','AQI','NOX','O3','CO','SO2', 'NO2', 'PM10', 'PM2_5']
    total_dist = sum(df_tmp.distance)
    
    for f in aqui_features:
        df_tmp[f]  = df_tmp[f]  * df_tmp['distance']
    
    for feature in aqui_features: 
        try:
            
            row[feature] = sum(df_tmp[feature])/ df_tmp[~df_tmp[feature].isnull()].distance.sum()
        except Exception as e:
            row[feature] = None
        
    return row
```

```python
def compute_distance(row, lat1, lon1):
    lat2 = float(row['latitude'])
    lon2 = float(row['longitude'])
    return gpxpy.geo.haversine_distance(lat1, lon1, lat2, lon2)


# In[10]:
def to_mongo(df):
    records = json.loads(df.T.to_json()).values()
    db_air_quality.weather_aqi_aggregation_hkust_5000_1800_subhour.insert(records)
```

```python
list_ = []

df_weather = df_weather.sort_values('time', ascending = True)
df_air_quality = df_air_quality.sort_values('time', ascending = True)
```

```python
count = 0 
for index, row in df_weather.iterrows(): 
    time_ =  row['time']
    lat = row['latitude']
    lon = row['longitude']
    
    df_tmp = df_air_quality.copy()
    df_tmp['time'] = abs(df_tmp['time'] - time_)
    filter_ = df_tmp.time <= time_diff
    
    df_tmp = df_tmp[filter_]
    df_tmp['distance'] = df_tmp.apply(lambda row : compute_distance(row, lat, lon ), axis=1)
    df_tmp = df_tmp[df_tmp.distance < distance_diff]
    
    row = interpolate_AQI(row, df_tmp)
    row['timestamp'] = row['time']
    row['time'] = datetime.datetime.fromtimestamp(int(row['time'])).strftime('%Y-%m-%d %H:%M:%S')
    
    # debug variables 
       
    row['max_distance'] = df_tmp['distance'].max()
    row['min_distance'] = df_tmp['distance'].min()
    row['total_rows'] =  len(df_tmp)
    
    row['max_time'] = df_tmp['time'].max()
    row['min_time'] = df_tmp['time'].min()
    
    list_.append(row)
    if count % 1001 == 1:
        print("Inserted" , count, len(list_))
        df_weather_augmented = pd.DataFrame(list_)
        to_mongo(df_weather_augmented)
        
        list_ = []

    count+=1
    


df_weather_augmented = pd.DataFrame(list_)

to_mongo(df_weather_augmented)
to_mongo(df_air_quality)
```
